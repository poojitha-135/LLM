```python
from transformers import pipeline
# Initialize sentiment analysis model
analyzer = pipeline(task="sentiment-analysis")
# Read input sentence from user
sentence = input("Enter a sentence for sentiment analysis: ")
# Analyze sentiment
output = analyzer(sentence)
# Get values from output
sentiment_label = output[0]["label"]
confidence_value = output[0]["score"]
# Print result
print("Sentiment:", sentiment_label)
print("Confidence Score:", round(confidence_value, 2))
```
