```python
#1.Whitespace Tokenization
text = input("Enter text: ")
tokens = text.split()
print("Whitespace Tokens:")
print(tokens)
```

```python
#2.Character-level Tokenization
text = input("Enter text: ")
char_tokens = list(text)
print("Character Tokens:")
print(char_tokens)
```

```python
#3.Subword Tokenization
!pip install transformers
from transformers import BertTokenizer
text = input("Enter text: ")
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
subword_tokens = tokenizer.tokenize(text)
print("Subword Tokens:")
print(subword_tokens)
```

```python
#4.Stop Word Removal
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Added to resolve LookupError
text = input("Enter text: ")
words = word_tokenize(text)
stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if w.lower() not in stop_words]
print("After Stop Word Removal:")
print(filtered_words)
```

```python
#5.Sentence & Word Tokenization

import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt')
text = input("Enter text: ")
sentences = sent_tokenize(text)
words = word_tokenize(text)
print("\nSentence Tokens:")
print(sentences)
print("\nWord Tokens:")
print(words)
```

```python
#6.Hugging Face (NLP Library)
from transformers import AutoTokenizer
text = input("Enter text: ")
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
tokens = tokenizer.tokenize(text)
print("Tokens:")
print(tokens)
```

```python
#1.One-Hot Encoding
text = input("Enter sentence: ")
words = text.split()
vocab = list(set(words))
one_hot = {}
for word in words:
    vector = [1 if v == word else 0 for v in vocab]
    one_hot[word] = vector
print("Vocabulary:", vocab)
print("One-Hot Encoding:")
for k, v in one_hot.items():
    print(k, ":", v)
```

```python
#2.Bag of Words
!pip install scikit-learn
from sklearn.feature_extraction.text import CountVectorizer
text = input("Enter sentence: ")
vectorizer = CountVectorizer()
bow = vectorizer.fit_transform([text])
print("Vocabulary:", vectorizer.get_feature_names_out())
print("Bag of Words Vector:", bow.toarray())
```

```python
#3.word2Vec
!pip install gensim
from gensim.models import Word2Vec
text = input("Enter sentence: ")
sentences = [text.split()]
model = Word2Vec(sentences, vector_size=50, window=5, min_count=1)
word = input("Enter word to get embedding: ")
print("Word2Vec Embedding:")
print(model.wv[word])
```

```python
#4.GloVe Embeddings
import gensim.downloader as api
# Load pretrained GloVe model
glove = api.load("glove-wiki-gigaword-50")
# Take input word
word = input("Enter a word: ").lower()
# Check if word exists in vocabulary
if word in glove:
 print("GloVe Embedding:")
 print(glove[word])
else:
 print("Word not found in GloVe vocabulary")
```

```python
#5.Bert Embedding
import torch
from transformers import BertTokenizer, BertModel
# Load BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")
# Take input sentence
text = input("Enter a text: ")
# Convert sentence into BERT input format
inputs = tokenizer(text, return_tensors="pt")
# Get model outputs
outputs = model(**inputs)
# Extract CLS token embedding
cls_embedding = outputs.last_hidden_state[:, 0, :]
# Display embedding
print("BERT CLS Embedding Shape:", cls_embedding.shape)
print("BERT Embedding:")
print(cls_embedding)
```

```python
import torch

# Input sentence
sentence = "every moment is a beginning"

# Tokenization (basic whitespace)
tokens = sentence.split()

# Embedding dimensions
embedding_dim = 8

# Create random embeddings for each token
embeddings = torch.randn(len(tokens), embedding_dim)

# Display token and its embedding
for token, emb in zip(tokens, embeddings):
    print(f"Token: {token}")
    print(f"Embedding: {emb}\n")
```
